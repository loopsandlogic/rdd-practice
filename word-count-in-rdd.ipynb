```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Spark SQL Example").getOrCreate()
```

```python
data = ["Pulastya Narain Pandey",
        "Arti Kumari Pandey",
        "Aprajita Narain Pandey",
        "Augustya Narayan Pandey",
        "Vashishth Narayan Pandey",
        "Tarkeshwar Pandey",
        "Kameshwar Pandey",
        "Nakuleshwar Pandey",
        "Akhileshwar Pandey",
        "Bishwamitra             Pandey",
        "      Rishabh.     Pandey"]

rdd = spark.sparkContext.parallelize(data)
rdd.collect()
```

```python
total_words = rdd.flatMap(lambda x: x.split()).count()
```

```python
print(f"Total number of words in the RDD: {total_words}")
```
